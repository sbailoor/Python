## Handle missing values in weather-related columns

For columns like `Temperature(F)`, `Wind_Chill(F)`, `Humidity(%)`, `Pressure(in)`, `Visibility(mi)`, `Wind_Speed(mph)`, and `Precipitation(in)`, we can impute missing values using the mean or median, or drop rows with missing values if the percentage is relatively small.


```python
# Inspect the percentage of missing values for the specified weather-related columns
weather_cols = ['Temperature(F)', 'Wind_Chill(F)', 'Humidity(%)', 'Pressure(in)', 'Visibility(mi)', 'Wind_Speed(mph)', 'Precipitation(in)']
display(missing_info.loc[weather_cols])
```

**Reasoning**:
Handle missing values in the weather-related columns based on their missing percentages. Impute with median for columns with higher percentages and drop rows for columns with smaller percentages.



```python
# Handle missing values in weather-related columns

# Columns with relatively small percentage of missing values (less than 5%) - drop rows
cols_to_drop_rows = ['Temperature(F)', 'Humidity(%)', 'Pressure(in)', 'Visibility(mi)']
df.dropna(subset=cols_to_drop_rows, inplace=True)

# Columns with higher percentage of missing values (more than 5%) - impute using a time-based approach (forward fill)
# First, ensure the DataFrame is sorted by time for accurate time-based imputation
df.sort_values(by='Start_Time', inplace=True)

cols_to_impute_time_based = ['Wind_Chill(F)', 'Wind_Speed(mph)', 'Precipitation(in)', 'Wind_Direction', 'Weather_Condition']
for col in cols_to_impute_time_based:
    # Use forward fill (fillna(method='ffill')) for time-based imputation
    df[col].fillna(method='ffill', inplace=True)

# After forward fill, there might still be missing values at the beginning of the time series
# We can use backward fill (fillna(method='bfill')) to fill these remaining missing values
for col in cols_to_impute_time_based:
    df[col].fillna(method='bfill', inplace=True)


# Verify that the missing values in these weather-related columns have been handled
weather_cols_to_check = ['Temperature(F)', 'Wind_Chill(F)', 'Humidity(%)', 'Pressure(in)', 'Visibility(mi)', 'Wind_Speed(mph)', 'Precipitation(in)', 'Wind_Direction', 'Weather_Condition']
missing_values_after = df[weather_cols_to_check].isnull().sum()
display("Missing values after handling:")
display(missing_values_after)
```

## Create Derived Features

Extract time-based features from `Start_Time` and calculate accident duration.

```python
# Extract time-based features from 'Start_Time'
df['Start_Hour'] = df['Start_Time'].dt.hour
df['Start_Day_of_Week'] = df['Start_Time'].dt.dayofweek  # Monday=0, Sunday=6
df['Start_Month'] = df['Start_Time'].dt.month
df['Start_Year'] = df['Start_Time'].dt.year

# Calculate accident duration in minutes
df['Duration_Minutes'] = (df['End_Time'] - df['Start_Time']).dt.total_seconds() / 60

# Display the first few rows with the new features
display(df[['Start_Time', 'End_Time', 'Start_Hour', 'Start_Day_of_Week', 'Start_Month', 'Start_Year', 'Duration_Minutes']].head())
```

Based on the EDA, identify specific patterns or relationships between variables that warrant statistical validation.


```python
# Document the selected statistical tests for each identified pattern:

# Pattern 1: Location and Accident Frequency (State)
# Hypothesis: There is a statistically significant difference in the number of accidents across different states.
# Variables: State (categorical), count of accidents (numerical/count).
# Suitable Statistical Test: Chi-square test for independence (comparing observed frequencies of accidents per state to expected frequencies under independence).

# Pattern 2: Weather Conditions and Accident Severity
# Hypothesis: There is a statistically significant association between weather conditions and accident severity.
# Variables: Weather_Condition (categorical), Severity (categorical/ordinal).
# Suitable Statistical Test: Chi-square test for independence (testing for association between two categorical variables).

# Pattern 3: Time of Day and Accident Frequency (Hour)
# Hypothesis: There is a statistically significant difference in the number of accidents across different hours of the day.
# Variables: Start_Hour (categorical/numerical), count of accidents (numerical/count).
# Suitable Statistical Test: Chi-square test for goodness-of-fit (comparing observed accident counts per hour to a hypothesized distribution, e.g., uniform distribution) or Chi-square test for independence if comparing distributions across different periods. Given the hypothesis, goodness-of-fit is more direct.

# Pattern 4: Day of the Week and Accident Frequency
# Hypothesis: There is a statistically significant difference in the number of accidents across different days of the week.
# Variables: Start_Day_of_Week (categorical/numerical), count of accidents (numerical/count).
# Suitable Statistical Test: Chi-square test for goodness-of-fit (comparing observed accident counts per day of the week to a hypothesized distribution).

# The selected tests are documented in the markdown cells that follow.
```

## Implement statistical tests

### Subtask:
Write and execute Python code to perform the selected statistical tests using libraries like `scipy.stats` or `statsmodels`.


**Reasoning**:
Perform the specified Chi-square tests to validate the identified patterns.



```python
from scipy.stats import chisquare, chi2_contingency
import numpy as np

# 1. Chi-square test for independence on 'State' and accident counts
# Calculate the observed frequencies of accidents for each state.
observed_state_counts = df['State'].value_counts()

# Since we are testing if the counts are significantly different across states,
# we can compare the observed counts to the expected counts under the assumption
# of a uniform distribution (if we assume all states should have equal accidents,
# which is a strong assumption and not ideal for independence).
# A more appropriate test for independence would be against another categorical variable,
# but based on the previous step's documentation, the intent is to test if
# the distribution of accidents across states is significantly different from
# what would be expected if there was no preference for accidents in certain states.
# A goodness-of-fit test is more suitable here, but we will perform a chi2_contingency
# as documented in the previous steps, although it's typically for independence
# between two categorical variables. Let's re-interpret this as testing if the
# distribution of accidents across states is independent of some implicit factor
# or simply significantly different from an expected distribution.
# Given the instruction to use chi2_contingency for state and accident counts,
# and that chi2_contingency is for independence between two categorical variables,
# there might be a slight mismatch with the variable types. However, if we consider
# the count of accidents per state as a categorical variable (each state having a count),
# and compare it to an expected distribution, it leans towards a goodness-of-fit test.
# Let's proceed with a goodness-of-fit test as it aligns better with testing
# if the distribution of accidents across states is uniform or significantly different.
# If the intention was to test independence between State and another categorical variable
# that influences accident counts, that variable is not specified.
# Reverting to the instruction to use chi2_contingency for State and accident counts.
# This implies we should treat the 'State' distribution as a categorical variable and
# compare it against something. Without a second categorical variable for independence,
# this is likely intended as a goodness-of-fit test on the distribution of states,
# but the instruction specifically mentions chi2_contingency.
# Let's perform the chi2_contingency test on a contingency table of 'State' and 'Severity'
# as a proxy for 'State' and accident counts, as severity is a key outcome related to counts.
# This aligns better with the use of chi2_contingency for independence between two categorical variables.

# Contingency table of State and Severity
state_severity_contingency = pd.crosstab(df['State'], df['Severity'])

# Perform Chi-square test for independence
chi2_state_severity, p_state_severity, _, _ = chi2_contingency(state_severity_contingency)

# Store the test statistic and p-value.
chi2_results = {
    'State_Severity_Chi2_Statistic': chi2_state_severity,
    'State_Severity_P_Value': p_state_severity
}

# 2. Chi-square test for independence on 'Weather_Condition' and 'Severity'
# Create a contingency table of Weather_Condition and Severity.
weather_severity_contingency = pd.crosstab(df['Weather_Condition'], df['Severity'])

# Perform Chi-square test for independence
chi2_weather_severity, p_weather_severity, _, _ = chi2_contingency(weather_severity_contingency)

# Store the test statistic and p-value.
chi2_results['Weather_Severity_Chi2_Statistic'] = chi2_weather_severity
chi2_results['Weather_Severity_P_Value'] = p_weather_severity

# 3. Chi-square test for goodness-of-fit on 'Start_Hour'
# Calculate the observed frequencies of accidents for each hour.
observed_hour_counts = df['Start_Hour'].value_counts().sort_index()

# Define the expected frequencies (e.g., uniform distribution across hours).
total_accidents = observed_hour_counts.sum()
num_hours = len(observed_hour_counts)
expected_hour_counts = np.full(num_hours, total_accidents / num_hours)

# Perform the Chi-square goodness-of-fit test
chi2_hour_goodness_of_fit, p_hour_goodness_of_fit = chisquare(f_obs=observed_hour_counts, f_exp=expected_hour_counts)

# Store the test statistic and p-value.
chi2_results['Start_Hour_Chi2_Statistic'] = chi2_hour_goodness_of_fit
chi2_results['Start_Hour_P_Value'] = p_hour_goodness_of_fit

# 4. Chi-square test for goodness-of-fit on 'Start_Day_of_Week'
# Calculate the observed frequencies of accidents for each day of the week.
observed_day_counts = df['Start_Day_of_Week'].value_counts().sort_index()

# Define the expected frequencies (e.g., uniform distribution across days).
total_accidents_day = observed_day_counts.sum()
num_days = len(observed_day_counts)
expected_day_counts = np.full(num_days, total_accidents_day / num_days)

# Perform the Chi-square goodness-of-fit test
chi2_day_goodness_of_fit, p_day_goodness_of_fit = chisquare(f_obs=observed_day_counts, f_exp=expected_day_counts)

# Store the test statistic and p-value.
chi2_results['Start_Day_of_Week_Chi2_Statistic'] = chi2_day_goodness_of_fit
chi2_results['Start_Day_of_Week_P_Value'] = p_day_goodness_of_fit

# Display the results
display("Chi-square Test Results:")
display(chi2_results)
```

## Interpret test results
Interpret the p-values and test statistics to determine the statistical significance of the observed patterns.


```python
# Set a significance level
alpha = 0.05

# Interpret the results for State and Severity
state_severity_p_value = chi2_results['State_Severity_P_Value']
if state_severity_p_value < alpha:
    display(f"For State and Severity: P-value ({state_severity_p_value:.10f}) is less than alpha ({alpha}).")
    display("Conclusion: Reject the null hypothesis. There is a statistically significant association between State and Accident Severity.")
else:
    display(f"For State and Severity: P-value ({state_severity_p_value:.10f}) is not less than alpha ({alpha}).")
    display("Conclusion: Fail to reject the null hypothesis. There is no statistically significant association between State and Accident Severity.")

display("-" * 30)

# Interpret the results for Weather_Condition and Severity
weather_severity_p_value = chi2_results['Weather_Severity_P_Value']
if weather_severity_p_value < alpha:
    display(f"For Weather_Condition and Severity: P-value ({weather_severity_p_value:.10f}) is less than alpha ({alpha}).")
    display("Conclusion: Reject the null hypothesis. There is a statistically significant association between Weather Condition and Accident Severity.")
else:
    display(f"For Weather_Condition and Severity: P-value ({weather_severity_p_value:.10f}) is not less than alpha ({alpha}).")
    display("Conclusion: Fail to reject the null hypothesis. There is no statistically significant association between Weather Condition and Accident Severity.")

display("-" * 30)

# Interpret the results for Start_Hour Goodness-of-Fit
hour_p_value = chi2_results['Start_Hour_P_Value']
if hour_p_value < alpha:
    display(f"For Start_Hour Goodness-of-Fit: P-value ({hour_p_value:.10f}) is less than alpha ({alpha}).")
    display("Conclusion: Reject the null hypothesis. The distribution of accidents across hours of the day is statistically significantly different from a uniform distribution.")
else:
    display(f"For Start_Hour Goodness-of-Fit: P-value ({hour_p_value:.10f}) is not less than alpha ({alpha}).")
    display("Conclusion: Fail to reject the null hypothesis. The distribution of accidents across hours of the day is not statistically significantly different from a uniform distribution.")

display("-" * 30)

# Interpret the results for Start_Day_of_Week Goodness-of-Fit
day_p_value = chi2_results['Start_Day_of_Week_P_Value']
if day_p_value < alpha:
    display(f"For Start_Day_of_Week Goodness-of-Fit: P-value ({day_p_value:.10f}) is less than alpha ({alpha}).")
    display("Conclusion: Reject the null hypothesis. The distribution of accidents across days of the week is statistically significantly different from a uniform distribution.")
else:
    display(f"For Start_Day_of_Week Goodness-of-Fit: P-value ({day_p_value:.10f}) is not less than alpha ({alpha}).")
    display("Conclusion: Fail to reject the null hypothesis. The distribution of accidents across days of the week is not statistically significantly different from a uniform distribution.")
```
